# WaveSync Timing Summary
Load CSVs from simulator or Flutter logs and compute basic stats.

```python
import pandas as pd
from pathlib import Path

# Configure your CSV paths
csv_paths = ["sim.csv", "deviceA.csv", "deviceB.csv"]
frames = []
for p in csv_paths:
    if Path(p).exists():
        df = pd.read_csv(p)
        df['source'] = p
        frames.append(df)

if not frames:
    raise SystemExit('No CSV files found. Place sim.csv/deviceA.csv etc in this folder.')

data = pd.concat(frames, ignore_index=True)

def describe_col(col):
    s = data[col].dropna()
    return {
        'count': int(s.count()),
        'mean': float(s.mean()),
        'median': float(s.median()),
        'stddev': float(s.std())
    }

for col in ['deltaFromTargetMs', 'drift']:
    if col in data.columns:
        print(col, describe_col(col))

data.head()
```
